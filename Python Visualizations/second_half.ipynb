```python
import mysql.connector
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches

# Establish a connection to your MySQL database
cnx = mysql.connector.connect(
    host='localhost',
    port=3306,
    user='root',
    password='Aycelen,12',
    database='schema'
)

# Query your database and retrieve the data using pandas
query = "SELECT iso_code, female_death, male_death FROM avg_10"


df = pd.read_sql_query(query, cnx)

# Close the MySQL connection
cnx.close()

# Create a horizontal bar chart using matplotlib
plt.figure(figsize=(10, 6))  # Set the figure size (width, height)

# Set the y-axis labels and corresponding values
y_labels = df['iso_code']
y_values = np.arange(len(df))

# Set the height of the bars
bar_height = 0.35

# Plot the combined deaths as stacked horizontal bars
plt.barh(y_values, df['female_death'], height=bar_height, align='center', label='Female Death', color='blue')
plt.barh(y_values, df['male_death'], height=bar_height, align='center', label='Male Death', color='orange', left=df['female_death'])

# Customize the chart
plt.xlabel('Distribution of Deaths by Sex ')
plt.ylabel('Countries')
plt.title('Top 10 Countries With the Most Lung Cancer Deaths Due to Smoking ')
plt.yticks(y_values, y_labels)  # Set the y-axis labels

# Create custom legend labels
blue_patch = mpatches.Patch(color='blue', label='Female')
orange_patch = mpatches.Patch(color='orange', label='Male')

# Add the legend to the chart
plt.legend(handles=[blue_patch, orange_patch], loc='upper right')


# Show the chart
plt.tight_layout()  # Adjust the layout to prevent overlapping labels
plt.show()
```

```python
import mysql.connector
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches

# Establish a connection to your MySQL database
cnx = mysql.connector.connect(
    host='localhost',
    port=3306,
    user='root',
    password='Aycelen,12',
    database='schema'
)

query = "SELECT iso_code, data_year, sales_daily FROM top10sales"
df = pd.read_sql_query(query, cnx)
cnx.close()
df_pivot = df.pivot(index='data_year', columns='iso_code', values='sales_daily')
plt.figure(figsize=(20, 12))  # Set the figure size (width, height)

# Plot the area chart
df_pivot.plot.area()

# Customize the chart
plt.xlabel('Years')
plt.ylabel('Daily Average Cigarette Sales')
plt.title('Top 10 Countries with Most Cigarette Sales')
plt.legend(title='Countries')


# Show the chart
plt.tight_layout()
plt.show()
```
